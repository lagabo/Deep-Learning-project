```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

dir = (Path.cwd() / ".." / "data" / "raw" / "bullflagdetector" / "Q5C0UO").resolve()
csv_paths = sorted(dir.glob("*.csv"))
print("data dir:", dir)
print("csv files:", len(csv_paths))

p = csv_paths[0]
df = pd.read_csv(p)
print("file:", p.name)
print("shape:", df.shape)
print("columns:", list(df.columns))
print(df.head(3))

plt.figure(figsize=(10, 3))
if "close" in df.columns:
    plt.plot(df["close"].values[:300])
    plt.title(f"close (first 300) - {p.name}")
else:
    plt.plot(df.iloc[:300, 0].values)
    plt.title(f"first column (first 300) - {p.name}")
plt.tight_layout()
plt.show()
```
